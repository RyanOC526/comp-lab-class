```python
import MDAnalysis as mda
import MDAnalysis.analysis.rdf as rdf
import MDAnalysis.transformations as transformations
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
os.chdir('/home/rmo302/compclass/comp-lab-class/Week12/Runs/Water')
```

```python
#Energy vs Time

time = []
energy = []

water = np.loadtxt("water_amid_1ps_every0.5fs-1.ener")
i = 0

while i<len(water):
    time.append(water[i][1])
    energy.append(water[i][4])
    i+=1

plt.title("Water AIMD Energy vs Time")
plt.plot(time, energy)
plt.xlabel("Time (fs)")
plt.ylabel("Potential Energy (a.u)")
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week12/Figures/Water PE vs Time .pdf')

plt.show()
```

```python
#g(r)

u = mda.Universe("water.xyz", "water_amid_1ps_every0.5fs-pos-1.xyz", topology_format = "XYZ")

#setting box dimensions
dim = [12.42, 12.42, 12.42, 90, 90, 90]
transform = transformations.boxdimensions.set_dimensions(dim)
u.trajectory.add_transformations(transform)

RDFOO = rdf.InterRDF(u.select_atoms("name O"), u.select_atoms("name O"), bins = 100, range = (1.5, 6.21))
RDFOO.run()
    


plt.plot(RDFOO.results.bins, RDFOO.results.rdf)


plt.xlabel('r')
plt.ylabel('g_OO(r)')
plt.xlim()
plt.ylim(-1,8)
plt.title('g_OO(r)')
plt.title('g_11(r)')
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week12/Figures/Water g_OO(r).pdf')

plt.show()
    
```

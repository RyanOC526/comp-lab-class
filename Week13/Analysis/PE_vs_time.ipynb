```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
os.chdir('/home/rmo302/compclass/comp-lab-class/Week13/Outputs')
```

```python
#NVT Energy vs Time

nvt_time = []
nvt_energy = []

nvt= np.loadtxt("NVT-1.ener")
i = 0

while i<len(nvt):
    nvt_time.append(nvt[i][1])
    nvt_energy.append(nvt[i][4])
    i+=1

plt.title("NVT Energy vs Time")
plt.plot(nvt_time, nvt_energy)
plt.xlabel("Time (fs)")
plt.ylabel("Potential Energy (a.u)")
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week13/Figures/NVT PE vs Time.pdf')

plt.show()
```

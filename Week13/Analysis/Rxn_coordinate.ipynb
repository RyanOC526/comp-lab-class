```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#Load Files

qm_time = []
qm_coordinate = []

md_time= []
md_coordinate = []

hills_time = []
hills_coordinate = []

os.chdir('/home/rmo302/compclass/comp-lab-class/Week13/Outputs/monitor')

qm_rxn= np.loadtxt("MONITOR-COLVAR.metadynLog")
i = 0

while i<len(qm_rxn):
    qm_time.append(qm_rxn[i][0])
    qm_coordinate.append(qm_rxn[i][1])
    i+=1

os.chdir('/home/rmo302/compclass/comp-lab-class/Week13/Outputs/metad1_initmonitor')

md_rxn = np.loadtxt("METADYN-COLVAR.metadynLog")
j = 0

while (j<len(md_rxn)):
    md_time.append(md_rxn[j][0])
    md_coordinate.append(md_rxn[j][1])
    j+=1

os.chdir('/home/rmo302/compclass/comp-lab-class/Week13/Outputs/metad1_initmonitor_freqhills')

hills_time = []
hills_coordinate = []

hills_rxn= np.loadtxt("METADYN-COLVAR.metadynLog")
k = 0

while k<len(hills_rxn):
    hills_time.append(hills_rxn[k][0])
    hills_coordinate.append(hills_rxn[k][1])
    k+=1

```

```python
#Reaction Coordinatae

plt.title("CC-CO Bond Distance vs Time")
plt.plot(qm_time, qm_coordinate, label = "QM/MM")
plt.plot(md_time, md_coordinate, label = "MD:H=1.5,W=0.5,F=100")
plt.plot(hills_time, hills_coordinate, label = "MD:H=1.5,W=0.5,F=20")
plt.xlabel("Time (fs)")
plt.ylabel("CC-CO Bond Distance (A)")
plt.legend()

plt.savefig('/home/rmo302/compclass/comp-lab-class/Week13/Figures/Rxn_Coordinate vs Time.pdf')

plt.show()
```

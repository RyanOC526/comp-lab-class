```python
# this notebook is for plotting
# the end to end analysis
import numpy as np
import matplotlib.pyplot as plt
#load mdtraj
import mdtraj as md
```

```python
import os
# change python directory to /home/gmh4/projects/CompLab/comp-lab-class
# top file is ../Data/1hz3_T310.start.nowater.gro
# trajectory file is ../Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc
# load the trajectory
traj = md.load_xtc('../Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='../Data/1hz3_T310.start.nowater.gro')
```

```python
# compute the distance between the first and last atom
def end_end(traj):
    # get the first and last atom
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atom
    dist = np.sqrt(np.sum((first-last)**2, axis=1))
    return dist

# compute the distance between the first and last atom
dist = end_end(traj)

#compute the radius of gyration at each time
radius = md.compute_rg(traj, masses = None)

#plot distance and radius of gyration vs time
plt.plot(traj.time, dist)
plt.plot (traj.time, radius, c= 'r')


# make this plot look nice
plt.xlabel('Time (ps)')
plt.ylabel('Distance (nm)')
plt.title('End to End Distance and Radius of Gyration')
plt.show()
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/End to End Distance and Radius of Gyration vs Time.pdf')

```

```python
#Normalized histrogram of End to End Distance

fig, ax = plt.subplots()

ax.hist(dist, label = "End to End Distance", histtype = 'step', density = True)
plt.legend()
plt.show()
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/End to End Distance Histogram.pdf')
```

```python
#Normalized historgram of Radius of Gyration

fig, ax = plt.subplots()
ax.hist(radius, label = "Radius of Gyration", histtype = 'step', density = True)
plt.legend()
plt.show()
plt. savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/Radius of Gyration Histogram.pdf')
```

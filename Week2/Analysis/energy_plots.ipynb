```python
#this notebook is for energy plots
import numpy as np
import matplotlib.pyplot as plt
```

```python
#change directory
import os
os.chdir('/home/rmo302/compclass/comp-lab-class/Week2/Data')
#read in file
f = open('1hz3_T310.run.25000000.energy.xvg')
#this will make arrays of Time, PE, KE, Total, Temp, Pressure and Vol
time = []
pot_ener = []
kin_ener = []
tot_ener = []
temp = []
press = []
vol = []    
```

```python
#puts elements from file into each array as floats
#splits each line, if the length is 7, which is the length of each row with numbers, then it will put each in array
#if the element is a comment in the beginnig, it clears the variable and moves on
for line in f:
    colm = line.split()
    if colm[0] == '#' or colm[0]=='@':
        colm.clear()
    elif len(colm) == 7:
        time.append(float(colm[0])/1000)
        pot_ener.append(float(colm[1]))
        kin_ener.append(float(colm[2]))
        tot_ener.append(float(colm[3]))
        temp.append(float(colm[4]))
        press.append(float(colm[5]))
        vol.append(float(colm[6]))
```

```python
#plots pot.ener. vs time
plt.plot (time, pot_ener, color = "r")
plt.xlabel("Time (ns)")
plt.ylabel("Potential Energy (kJ/mol)")
plt.title("Potential Energy vs Time")

plt.show()
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/Potential Energy vs Time.png')
```

```python
#plots kin.ener. vs time
plt.plot (time, kin_ener, color = "b")
plt.xlabel("Time (ns)")
plt.ylabel("Kinetic Energy (kJ/mol)")
plt.title("Kinetic Energy vs Time")

plt.show()
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/Kinetic Energy vs Time.pdf')
```

```python
#plots tot.ener. vs time
plt.plot (time, tot_ener)
plt.xlabel("Time (ns)")
plt.ylabel("Total Energy (kJ/mol)")
plt.title("Total Energy vs Time")

plt.show()
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/Total Energy vs Time.pdf')
```

```python
#plots temp vs time
plt.plot (time, temp, color = "g")
plt.xlabel("Time (ns)")
plt.ylabel("Temperature (K)")
plt.title("Temperature vs Time")

plt.show()
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/Temperature vs Time.pdf')
```

```python
#plots press vs time
plt.plot (time, press, color = "m")
plt.xlabel("Time (ns)")
plt.ylabel("Pressure (bar)")
plt.title("Pressure vs Time")

plt.show()
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/Pressure vs Time.pdf')
```

```python
#plots vol vs time
plt.plot (time, vol, color = 'k')
plt.xlabel("Time (ns)")
plt.ylabel("Volume (nm^3)")
plt.title("Volume vs Time")

plt.show()
plt.savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/Volume vs Time.pdf')
```

```python
#plots temp and kin.ener vs time
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(time, temp, '-', color = "b")
ax2.plot(time, kin_ener, '--', color = "r")
ax1.set_xlabel ("Time (ns)")
ax1.set_ylabel ("Temperature (K)", color = "b")
ax2.set_ylabel ("Kinetic Energy (kJ/mol)", color = "r")
plt.title ("Temperature and Kinetic Energy vs Time")
plt.show()

plt.savefig('/home/rmo302/compclass/comp-lab-class/Week2/Figures/Temperatue and Kinetic Energy vs Time.pdf')
```

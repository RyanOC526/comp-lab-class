```python
import numpy as np
import matplotlib.pyplot as plt
import os 

os.chdir('/home/rmo302/compclass/comp-lab-class/Week3/Data')

x,y = np.loadtxt('pressure.xvg', comments = "#", unpack = True)
avg_y = []
window = 10
for ind in range (len(y) - window + 1):
    avg_y.append(np.mean(y[ind - window]))

for ind in range(window-1):
    avg_y.insert(0, np.nan)

plt.plot(x,y)
plt.plot(x, avg_y, c='r', label="10 ps Running Average",linewidth = 0.5)

plt.title("Pressure: 1AKI, NPT Equilibration")
plt.xlabel("Time (ps)")
plt.ylabel("Pressure (bar)")

plt.legend()

plt.show()
```

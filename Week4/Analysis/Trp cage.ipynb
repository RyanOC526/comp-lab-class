```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import os
```

```python
os.chdir("/home/rmo302/compclass/comp-lab-class/Week4/Setup/Protein/Trp_cage_AMBER/gromacs")

traj = md.load_xtc('step5_center.xtc', top = "step5_center.gro")
```

```python
# compute the distance between the first and last atom
def end_end(traj):
    # get the first and last atom
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atom
    dist = np.sqrt(np.sum((first-last)**2, axis=1))
    return dist

# compute the distance between the first and last atom
dist = end_end(traj)

#compute the radius of gyration at each time
radius = md.compute_rg(traj, masses = None)

#plot distance and radius of gyrations vs time 
plt.plot(traj.time, dist, label = 'End to End Distance')
plt.plot(traj.time, radius, c = 'r', label = 'Radius of Gyration')

#make this plot look nice
plt.xlabel ('Time (ps)')
plt.ylabel ('Distance (nm)')
plt.title ('End to End Distance and Radius of Gyration')

plt.legend()
plt.show()

plt.savefig('/home/rmo302/compclass/comp-lab-class/Week4/Figures/End to End Distance and Radius of Gyration.pdf')
```

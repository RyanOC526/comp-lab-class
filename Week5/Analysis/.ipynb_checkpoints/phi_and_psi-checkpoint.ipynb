```python
import numpy as np
import matplotlib.pyplot as plt
import os
import mdtraj as md
```

```python
#computes phi and psi dihedral angles

os.chdir('/home/rmo302/compclass/comp-lab-class/Week5/MD')

traj = md.load("adp_pt1.trr", top = "adp_pt1.gro")

phi = md.compute_phi(traj, periodic = True, opt = True)
psi = md.compute_psi(traj, periodic = True, opt = True)
```

```python
#plots phi versus time

plt.title("Phi vs Time")
plt.xlabel("Time (ps)")
plt.ylabel("Phi (Radians)")
plt.plot(traj.time, phi[1][0:])
```

```python
#plots psi vs time

plt.title("Psi vs Time")
plt.xlabel("Time (ps)")
plt.ylabel("Psi (Radians)")
plt.plot(traj.time, psi[1][0:])
```

```python
#2D histogram

plt.title("Ramachandran Plot")
plt.xlabel('Phi (Radians)')
plt.ylabel('Psi (Radians)')
plt.hist2d(phi[1][:,0], psi[1][:,0], bins = 40)
plt.colorbar()
plt.show()
```

```python
hist = np.histogram2d(phi[1][:,0], psi[1][:,0])

energy = -np.log(hist[0].T)

plt.imshow(energy, extent = [hist[1][0], hist[1][-1], hist[2][0], hist[2][-1]], 
           origin = 'lower')
plt.colorbar()
plt.title("Free Energy Surface")
plt.xlabel('Phi (Radians)')
plt.ylabel('Psi (Radians)')
```

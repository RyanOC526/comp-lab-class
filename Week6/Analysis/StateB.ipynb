```python
import numpy as np
import mdtraj as md
import matplotlib.pyplot as plt
import os
```

```python
# Read in COLVAR file for State B

os.chdir('/home/rmo302/compclass/comp-lab-class/Week6/masterclass-21-4/Exercise-1/StateB')

colvar = np.loadtxt("COLVAR")

time =[] #ps
phi_col=[]
psi_col =[]

i = 0
while i<len(colvar):
    time.append(colvar[i][0])
    phi_col.append(colvar[i][1])
    psi_col.append(colvar[i][2])
    i+=1
```

```python
#compute phi and psi for State B from xtc file

traj = md.load_xtc("traj_comp.xtc", top = "confout.gro")

phi_xtc = md.compute_phi(traj, periodic = True, opt = True)
psi_xtc = md.compute_psi(traj, periodic = True, opt = True)
```

```python
#plot phi vs psi from COLVAR and xtc

plt.title ("Phi vs Psi: State B")
plt.xlabel("Psi (Radians)")
plt.ylabel("Phi (Radians)")
plt.scatter(psi_col[::500], phi_col[::500], label = 'COLVAR')
plt.scatter(psi_xtc[1][0:][::500], phi_xtc[1][0:][::500], label = 'xtc', marker = ".", s = 30)
plt.legend()
plt.show()

plt.savefig('/home/rmo302/compclass/comp-lab-class/Week6/Figures/StateB Phi vs Psi.pdf')
```

```python
#compute RMSD of phi and psi for StateB

phi_sd = np.std(phi_col)
psi_sd = np.std(psi_col)

print (phi_sd)
print (psi_sd)
```

```python
import numpy as np
import mdtraj as md
import matplotlib.pyplot as plt
import os
```

```python
# Read in COLVAR file from metad

os.chdir('/home/rmo302/compclass/comp-lab-class/Week6/lugano-3')

colvar = np.loadtxt("COLVAR")

time_col =[] #ps
phi_col=[]
psi_col=[]
bias_col=[]

i = 0
while i<len(colvar):
    time_col.append(colvar[i][0]/1000)
    phi_col.append(colvar[i][1])
    psi_col.append(colvar[i][2])
    bias_col.append(colvar[i][3])
    i+=1
```

```python
plt.plot(time_col[::500], phi_col[::500])

plt.title("Phi vs Time: Metadynamics (SIGMA = 0.35)")
plt.xlabel ("Time (ns)")
plt.ylabel ("Phi (radians)")
plt.show()

plt.savefig('/home/rmo302/compclass/comp-lab-class/Week6/Figures/Phi vs Time.pdf')
```

```python
#read in HILLS file from metad

hills = np.loadtxt("HILLS")

time_hills=[]
phi_hills=[]
sigma=[]
height=[]
biasf=[]

i = 0
while i<len(hills):
    time_hills.append((hills[i][0])/1000)
    phi_hills.append(hills[i][1])
    sigma.append(hills[i][2])
    height.append(hills[i][3])
    biasf.append(hills[i][4])
    i+=1
```

```python
#plot hill height vs time

plt.scatter(time_hills, height)

plt.title("Hill Height vs Time")
plt.xlabel("Time (ns)")
plt.ylabel("Gaussian Hill Height (kJ/mol)")

plt.show()

plt.savefig('/home/rmo302/compclass/comp-lab-class/Week6/Figures/Hill Height vs Time.pdf')
```

```python
#read in Free Energy Surface 

fes = np.loadtxt("fes.dat")

phi_fes=[]
free_energy=[]

i = 0
while i<len(fes):
    phi_fes.append(fes[i][0])
    free_energy.append(fes[i][1])
    i+=1
```

```python
#plot free energy

plt.plot(phi_fes, free_energy)

plt.title("Free Energy vs Phi")
plt.xlabel("Phi (radians)")
plt.ylabel("Free Energy (kJ/mol)")

plt.show()

plt.savefig('/home/rmo302/compclass/comp-lab-class/Week6/Figures/Free Energy vs Phi.pdf')
```

```python
#read in dat files for every 100 Gaussians

phi_100=[]

energy1=[]
energy2=[]
energy3=[]
energy4=[]
energy5=[]
energy6=[]

plot1 = np.loadtxt("fes_1.dat") #0.1 ns
plot2 = np.loadtxt("fes_3.dat")#0.3 ns
plot3 = np.loadtxt("fes_10.dat")#1 ns
plot4 = np.loadtxt("fes_20.dat")#2 ns
plot5 = np.loadtxt("fes_50.dat")#5 ns
plot6 = np.loadtxt("fes_100.dat")# 10 ns

i = 0
while i<len(plot1):
    
    phi_100.append(plot1[i][0])
    
    energy1.append(plot1[i][1])
    energy2.append(plot2[i][1])
    energy3.append(plot3[i][1])
    energy4.append(plot4[i][1])
    energy5.append(plot5[i][1])
    energy6.append(plot6[i][1])
    i+=1
```

```python
#plotting every 100 Gaussians
plt.plot(phi_100, energy1, label = "0.1 ns")
plt.plot(phi_100, energy2, label = "0.3 ns")
plt.plot(phi_100, energy3, label = "1 ns")
plt.plot(phi_100, energy4, label = "2 ns")
plt.plot(phi_100, energy5, label = "5 ns")
plt.plot(phi_100, energy6, label = "10 ns")

plt.title("Free Energy vs Phi Every 100 Gaussians")
plt.xlabel("Phi (radians)")
plt.ylabel("Free Energy (kJ/mol)")
plt.legend()

plt.show()

plt.savefig('/home/rmo302/compclass/comp-lab-class/Week6/Figures/Free Energy vs Phi Every 100 Gaussians.pdf')
```

```python
import numpy as np
import mdtraj as md
import matplotlib.pyplot as plt
import os
```

```python
#load in the crystal structure

os.chdir('/home/rmo302/compclass/comp-lab-class/Week7')

crystal = md.load("3ry2.pdb")
crystal_top = crystal.topology
```

```python
#load in alphafold

os.chdir('/home/rmo302/compclass/comp-lab-class/Week7/3RY2_AlphaFold/')

alphafold = md.load("3ry2_alphafold_76ce1_unrelaxed_rank_1_model_3.pdb")
alpha_top = alphafold.topology
```

```python
#load in alphafold greene

os.chdir('/home/rmo302/compclass/comp-lab-class/Week7/3RY2')

greene1 = md.load("unrelaxed_model_1_pred_0.pdb")
greene2 = md.load("unrelaxed_model_2_pred_0.pdb")
greene3 = md.load("unrelaxed_model_3_pred_0.pdb")
greene4 = md.load("unrelaxed_model_4_pred_0.pdb")
greene5 = md.load("unrelaxed_model_5_pred_0.pdb")

greene1_top = greene1.topology
greene2_top = greene2.topology
greene3_top = greene3.topology
greene4_top = greene4.topology
greene5_top = greene5.topology

CA_values = []
Backbone_values = []
Sidechain_values = []
```

```python
#computing RMSD for Alphafold

alpha_CA = alpha_top.select("name == CA")
alpha_backbone = alpha_top.select("backbone")
alpha_sidechain = alpha_top.select("sidechain")

alphaCA_rmsd = md.rmsd(alphafold, crystal, atom_indices = alpha_CA, parallel=True, precentered=False)
alphaBackbone_rmsd = md.rmsd(alphafold, crystal, atom_indices = alpha_backbone, parallel=True, precentered=False)
alphaSidechain_rmsd = md.rmsd(alphafold, crystal, atom_indices = alpha_sidechain, parallel=True, precentered=False)

print(alphaCA_rmsd)
print(alphaBackbone_rmsd)
print(alphaSidechain_rmsd)
```

```python
#computing RMSD for Greene Model 1

greene1_CA = greene1_top.select("name ==CA")
greene1_backbone = greene1_top.select("backbone")
greene1_sidechain = greene1_top.select("sidechain")

greene1CA_rmsd = md.rmsd(greene1, crystal, atom_indices = greene1_CA, parallel = True, precentered = False)
greene1Backbone_rmsd = md.rmsd(greene1, crystal, atom_indices = greene1_backbone, parallel = True, precentered = False)
greene1Sidechain_rmsd = md.rmsd(greene1, crystal, atom_indices = greene1_sidechain, parallel = True, precentered = False)

print(greene1CA_rmsd)
print(greene1Backbone_rmsd)
print(greene1Sidechain_rmsd)
```

```python
#Bar Plot of Alpha-Carbon RMSD 

CA_values = [greene1CA_rmsd[0], greene2CA_rmsd[0], greene3CA_rmsd[0], greene4CA_rmsd[0], greene5CA_rmsd[0]]
x = [1, 2, 3, 4, 5]

plt.bar(x, CA_values)
plt.title("RMSD of Alpha Carbons")
plt.xlabel("Model")
plt.ylabel("RMSD (nm)")

plt.ylim(0.88, 0.95)

plt.savefig('/home/rmo302/compclass/comp-lab-class/Week7/Figures/RMSD of Alpha Carbons.pdf')
```
